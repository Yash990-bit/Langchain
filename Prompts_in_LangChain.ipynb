# Prompts
A prompt is simply the instruction or question you give to an AI.

# Formula for a Good Prompt

Role + Task + Context + Output Format

# Static Prompts vs Dynamic Prompts

1. Static Prompt

A static prompt is written once and stays the same for every request.

# Use Cases
Chatbots with fixed behavior

Customer support bots

Simple AI applications

# 2. Dynamic Prompt

A dynamic prompt changes based on the user, context, data, or previous conversation.

# Template:

You are a coding mentor.

User Name: {name}
Skill Level: {level}

Explain Binary Search accordingly.

If:

name = Yash

level = Beginner

Prompt becomes:

You are a coding mentor.

User Name: Yash

Skill Level: Beginner

Explain Binary Search accordingly.

For another user:

User Name: Rahul

Skill Level: Advanced

The prompt changes automatically.

# 📄 Static Prompt = Pre-written template (fixed)
#🔄 Dynamic Prompt = Template + Variables + Context (changes at runtime)

# Code

```python
# Static Prompt

prompt = """
You are a Python teacher.
Explain Binary Search in simple words.
"""

print(prompt)
```

```python
# Dynamic

topic = input("Enter topic: ")

prompt = f"""
You are a Python teacher.
Explain {topic} in simple words with examples.
"""

print(prompt)
```

# Prompt Template
A Prompt Template is a predefined format with empty placeholders that are filled with data later.

Think of it like a form.

* Normal Prompt

Explain Binary Search in simple words.

This is fixed.

# Prompt Template

Explain {topic} in simple words.

Here, {topic} is a placeholder.

When you provide a value:

topic = "Binary Search"

prompt = f"Explain {topic} in simple words."

* Final prompt:

Explain Binary Search in simple words.

# Why Use Prompt Templates?

Without template:

prompt1 = "Explain Python"

prompt2 = "Explain Java"

prompt3 = "Explain C++"

With template:

template = "Explain {topic}"

topic = "Python"

Just change the value, not the whole prompt.

# why use prompt template over f string

1. Reusability

f-string

topic = "Docker"

prompt = f"Explain {topic}"

topic = "Python"

prompt = f"Explain {topic}"

You keep recreating the prompt.

Template
template = "Explain {topic}"

template.format(topic="Docker")

template.format(topic="Python")

template.format(topic="React")

One template, many inputs.

2. User Input Comes Later

Imagine building a chatbot.

At startup:

template = """

Answer the question:
{question}
"""

Later when the user asks:

question = "What is Docker?"

prompt = template.format(question=question)

The template exists before the value arrives.

3. Multiple Variables

template = """

Explain {topic}

for a {level} student

using {examples} examples.
"""

Then:

template.format(
    topic="Binary Search",
    level="beginner",
    examples=3
)

Much cleaner for large prompts

# Message Placeholder

A message placeholder is a variable inside a chat message template that gets replaced with real data later.

# Simple Idea
Template (with placeholder)

User: Explain {topic} in simple words.

Here:

{topic} = message placeholder

# After filling value
User: Explain Docker in simple words.

# In Chat Prompt Template (Role-based)
[
  ("system", "You are a teacher."),
  ("human", "Explain {topic} in simple words.")
]

Here:

{topic} = message placeholder inside the human message

# Where Message Placeholders are Used
1. Chatbots

User: {question}

2. AI Assistants

System: You are a helpful assistant.

User: Answer {query}

3. RAG (Retrieval systems)

Context: {context}

Question: {question}

# Message placeholder is a variable (like {topic}, {question}) used inside chat messages that gets replaced with actual values at runtime to generate dynamic prompts for LLMs.
