# Langchain

LangChain is an open-source framework used to build applications powered by Large Language Models (LLMs) like GPT, Claude, Gemini, and others.

LangChain helps you connect an AI model with:

Documents (PDFs, CSVs, databases)

APIs

Tools

Memory

Agents

# Why LangChain?

# Without LangChain:

from openai import OpenAI

client = OpenAI()

response = client.chat.completions.create(
    model="gpt-4o",
    messages=[{"role": "user", "content": "What is AI?"}]
)

print(response.choices[0].message.content)

# With LangChain:

from langchain_openai import ChatOpenAI

llm = ChatOpenAI(model="gpt-4o")

response = llm.invoke("What is AI?")

print(response.content)

LangChain provides many extra features beyond just calling the model.

# Concept of Chain
A Chain is simply a sequence of steps where the output of one step becomes the input of the next step.



# Basic Chain Example


```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate

llm = ChatOpenAI()

prompt = ChatPromptTemplate.from_template(
    "Explain {topic}"
)

chain = prompt | llm

result = chain.invoke(
    {"topic": "Docker"}
)

print(result.content)
```
