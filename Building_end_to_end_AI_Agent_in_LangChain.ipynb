# ReACt

ReAct is a prompting framework used in AI agents where the model combines reasoning and taking actions to solve a task.

# Full form:

ReAct = Reasoning + Acting

The agent does not directly answer. It follows a loop:

Thought → Action → Observation → Thought → Action → Final Answer

# Why ReAct is needed?

A normal LLM:

User Question
      ↓
     LLM
      ↓
  Final Answer

Problem:

Cannot access external information

Cannot use tools

May hallucinate

Cannot perform multi-step tasks

ReAct adds the ability to:

Think about the problem

Decide what action is needed

Use tools

Observe results

Continue reasoning

```python
# ReAct Architecture

                 User
                  |
                  ↓
              AI Agent
                  |
                  ↓
                 LLM
                  |
        ---------------------
        |                   |
     Reason              Act
        |                   |
        ↓                   ↓
   Decide step        Call Tool
                            |
                            ↓
                      Tool Result
                            |
                            ↓
                       Observation
                            |
                            ↓
                     Back to LLM
```

# Agent and Agent Executor

Agent = Brain (decides what to do)

Agent Executor = Manager (runs the agent and executes actions)



# 1. What is an Agent?

An Agent is the reasoning component that decides:

What action should be taken?

Which tool should be used?

What input should be given to the tool?

Is the task completed or not?

The agent does not execute tools itself.

Example:

User:

"Calculate the square root of 144"

Agent thinks:

I need a calculator.

Use calculator tool.

Input: sqrt(144)

Agent output:

{
  "tool": "calculator",
  "input": "sqrt(144)"
}

It only suggests the action.

# 2. What is Agent Executor?

The Agent Executor is the runtime that actually runs the agent.

Its job:

Send user input to agent

Get agent decision

Execute the selected tool

Give tool output back to agent

Repeat the loop

Return final answer

Flow:

User
 |
 ↓
Agent Executor
 |
 ↓
Agent (LLM reasoning)
 |
 ↓

 |
 ↓

 |

 |
 ↓

 |
 ↓

 |
 ↓
