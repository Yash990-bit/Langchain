# Define

A language model is basically a computer program that can understand and generate human language

# How it works (idea level):

It has read a huge amount of text from books, websites, and articles. From that, it learns:

which words usually come together

how sentences are formed

how people normally talk or write

#  Statistical vs Modern Language Models

👉 This is a classification based on technology / approach

Easy explanation:

Ye models sirf word patterns count karke predict karte the.

Soch lo:

Agar “I am going to” ke baad log aksar “school” bolte hain

👉 to model wahi guess karega

# How it works:
It looks at previous words only
Uses simple probability (chance)

No real understanding

* Limitation:
Smart nahi hote

Long sentences samajh nahi pate

Context weak hota hai

👉 Example: N-gram models

# Neural / Modern Language Models (AI-based)

👉 “Smart deep learning models (like ChatGPT)”

Easy explanation:

Ye models brain jaisa network use karte hain jo language ko deeper level pe samajhta hai.

Soch lo:

Ye sirf words nahi dekhta

Ye full sentence + meaning + context samajhta hai
# How it works:
Deep learning (neural networks) use karta hai
Previous + next context dono consider karta hai
Large data se train hota hai
* Strength:
Very smart responses deta hai

Long conversations samajh leta hai

Human-like text generate karta hai

👉 Example: GPT, BERT, LLaMA

# 1. LLM (Large Language Models)

👉 “Very big AI models trained on huge data”

Easy Hinglish:

Ye models bahut zyada data (internet, books, code) se trained hote hain.

Bahut powerful hote hain

Long context samajh lete hain

Complex answers de sakte hain

👉 Example:

GPT models

LLaMA

Claude

📌 Simple idea:

“Ek super-smart brain jisko duniya bhar ka text padhaya gaya hai”

2. Chat Models (Instruction-tuned models)

👉 “LLM ka friendly version jo chat ke liye banaya gaya hai”

Easy Hinglish:

Ye basically LLM hi hota hai, but isko specially train kiya jata hai ki wo insaan jaisa baat kare

Questions ka direct answer deta hai

Conversation follow karta hai

Polite aur helpful hota hai

👉 Example:

ChatGPT

Claude Chat

Gemini chat mode

📌 Simple idea:

“LLM + training for chatting = Chat Model”

# Open Source Models
Open source models wo AI models hote hain jinka code + trained weights publicly available hote hain,

so anyone can:

download,
use,
modify,
improve

👉 Matlab: “free to use and study (with some license rules)”

* Closed model (GPT-4 etc.) = locked car 🚗 (sirf company drive kar sakti hai)
* Open source model = open bike 🏍️ (tum bhi use + modify kar sakte ho)

# Key features of open source models:
✅ Free access (mostly)

✅ Can run locally (your laptop/server)

✅ Can fine-tune for your project

❌ Sometimes less powerful than closed models

❌ Need setup + GPU for best performance

# LLM Code

```python
from langchain_openai import OpenAI
from dotenv import load_dotenv

load_dotenv()

llm = OpenAI(model='gpt-3.5-turbo-instruct')

result = llm.invoke("What is the capital of India")

print(result)
```

# ChatModels Code

```python
# openai

from langchain_openai import ChatOpenAI
from dotenv import load_dotenv

load_dotenv()

model = ChatOpenAI(model='gpt-4', temperature=1.5, max_completion_tokens=10)

result = model.invoke("Write a 5 line poem on cricket")

print(result.content)
```

```python
# anthropic
from langchain_anthropic import ChatAnthropic
from dotenv import load_dotenv

load_dotenv()

model = ChatAnthropic(model='claude-3-5-sonnet-20241022')

result = model.invoke('What is the capital of India')

print(result.content)
```

```python
# google
from langchain_google_genai import ChatGoogleGenerativeAI
from dotenv import load_dotenv

load_dotenv()

model = ChatGoogleGenerativeAI(model='gemini-1.5-pro')

result = model.invoke('What is the capital of India')

print(result.content)
```

```python
# hugging Face api
from langchain_huggingface import ChatHuggingFace, HuggingFaceEndpoint
from dotenv import load_dotenv

load_dotenv()

llm = HuggingFaceEndpoint(
    repo_id="TinyLlama/TinyLlama-1.1B-Chat-v1.0",
    task="text-generation"
)

model = ChatHuggingFace(llm=llm)

result = model.invoke("What is the capital of India")

print(result.content)
```

```python
# Hugging face local
from langchain_huggingface import ChatHuggingFace, HuggingFacePipeline
import os

os.environ['HF_HOME'] = 'D:/huggingface_cache'

llm = HuggingFacePipeline.from_model_id(
    model_id='TinyLlama/TinyLlama-1.1B-Chat-v1.0',
    task='text-generation',
    pipeline_kwargs=dict(
        temperature=0.5,
        max_new_tokens=100
    )
)
model = ChatHuggingFace(llm=llm)

result = model.invoke("What is the capital of India")

print(result.content)
```

# Embedding Model Code


```python
# open ai
from langchain_openai import OpenAIEmbeddings
from dotenv import load_dotenv

load_dotenv()

embedding = OpenAIEmbeddings(model='text-embedding-3-large', dimensions=32)

result = embedding.embed_query("Delhi is the capital of India")

print(str(result))
```

```python
# open ai embedding
from langchain_openai import OpenAIEmbeddings
from dotenv import load_dotenv

load_dotenv()

embedding = OpenAIEmbeddings(model='text-embedding-3-large', dimensions=32)

documents = [
    "Delhi is the capital of India",
    "Kolkata is the capital of West Bengal",
    "Paris is the capital of France"
]

result = embedding.embed_documents(documents)

print(str(result))
```

```python
# Document Similarity
from langchain_openai import OpenAIEmbeddings
from dotenv import load_dotenv
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

load_dotenv()

embedding = OpenAIEmbeddings(model='text-embedding-3-large', dimensions=300)

documents = [
    "Virat Kohli is an Indian cricketer known for his aggressive batting and leadership.",
    "MS Dhoni is a former Indian captain famous for his calm demeanor and finishing skills.",
    "Sachin Tendulkar, also known as the 'God of Cricket', holds many batting records.",
    "Rohit Sharma is known for his elegant batting and record-breaking double centuries.",
    "Jasprit Bumrah is an Indian fast bowler known for his unorthodox action and yorkers."
]

query = 'tell me about bumrah'

doc_embeddings = embedding.embed_documents(documents)
query_embedding = embedding.embed_query(query)

scores = cosine_similarity([query_embedding], doc_embeddings)[0]

index, score = sorted(list(enumerate(scores)),key=lambda x:x[1])[-1]

print(query)
print(documents[index])
print("similarity score is:", score)
```
