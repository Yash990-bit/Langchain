# Prompts
A prompt is just the instruction you give to an AI model.

It tells the AI:

what to do

how to answer

what format to follow

# Instead of writing raw text every time, LangChain uses Prompt Templates.

# Without LangChain
prompt = "Explain " + topic
# With LangChain
from langchain_core.prompts import

ChatPromptTemplate

prompt = ChatPromptTemplate.from_template(
    "Explain {topic} in simple words"
)

Now you can reuse it:

prompt.format(topic="Docker")

🧩 Types of Prompts
1. Simple Prompt

Just a direct question:

What is AI?

2. Instruction Prompt

You give rules:

Explain AI in simple words for a 10-year-old.

3. Role-based Prompt

You assign a role:

You are a Python teacher. Explain loops.

4. Structured Prompt

You define format:

Explain Docker in:

- Definition

- Features

- Use cases

# Chains

A Chain means connecting multiple steps together.

Flow:

User Input
 ↓
Prompt
 ↓
LLM
 ↓
Output

Each step's output becomes the next step's input.

This is called a Chain.

# Why Do We Need Chains?

# Without a chain:

Question
 ↓
AI
 ↓
Answer

For simple tasks, that's enough.

# But real applications need multiple steps:

Question
 ↓
Search Documents
 ↓
Get Relevant Data
 ↓
Send to AI
 ↓
Generate Answer

A chain manages all these steps.


# Example

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate

llm = ChatOpenAI()

prompt = ChatPromptTemplate.from_template(
    "Explain {topic}"
)

chain = prompt | llm

result = chain.invoke(
    {"topic": "Docker"}
)

print(result.content)
```

# Chain ka matlab hai multiple steps ko connect karna, jahan har step ka output agle step ka input ban jata hai

# Memory

Memory means the AI can remember previous conversations and use them in future responses.

# Without memory:

User: My name is Yash

AI: Nice to meet you.

User: What is my name?

AI: I don't know.

The AI forgot everything.

# With Memory
User: My name is Yash

AI: Nice to meet you.

User: What is my name?

AI: Your name is Yash.

The AI remembered.

# Memory Flow
User Message
      ↓
Store in Memory
      ↓
Next Question
      ↓
Read Memory
      ↓
Generate Answer

# Types of Memory
1. Conversation Buffer Memory

Stores the entire conversation.

Q1,
Q2,
Q3,
Q4

Everything is remembered.

2. Conversation Summary Memory

Instead of storing everything:

Long Conversation
      ↓
Short Summary

Saves tokens and cost.

3. Vector Memory

Stores important information as embeddings.

Conversation
 ↓
Embeddings
 ↓
Vector Store

Useful for long-term memory.

# What does "LLM API calls are stateless" mean?

The LLM does not remember previous requests automatically.

Every API call is treated as a completely new conversation.

# Why?

Because the API works like:

Request
 ↓
Model
 ↓
Response

After sending the response:

Memory = Deleted

The model doesn't keep a copy of your conversation.

# LLM API calls are stateless, meaning the model does not automatically remember previous interactions. Each request is independent, so conversation history must be stored and sent by the application if context is needed.

# Agents

An Agent is an AI that can think, decide, and choose which tool to use to complete a task.

# Normal LLM:

Question
 ↓
LLM
 ↓
Answer

# Agent:

Question
 ↓
Think
 ↓
Choose Tool
 ↓
Use Tool
 ↓
Answer

# Real-Life Example

Imagine you're a manager.

Employee asks:

What is 256 × 789?

You think:

I should use a calculator.

Then:

Calculator
 ↓
Result
 ↓
Answer

This decision-making ability is what an Agent does.

 ↓

 ↓

# An Agent is an AI system that uses an LLM to reason about a task, decide which tools to use, execute those tools, and generate a final response.
