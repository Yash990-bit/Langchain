# StrOutputParser

ka use tab hota hai jab hum LLM se output ko as-it-is plain text (string) form me lena chahte hain — without any structure or JSON conversion.

# LLM output ko bas clean string ke form me pass karna

# Code


```python
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI

# LLM
llm = ChatOpenAI(model="gpt-4o-mini")

# Prompt
prompt = ChatPromptTemplate.from_template(
    "Explain {topic} in simple words"
)

# String Output Parser
parser = StrOutputParser()

# Chain
chain = prompt | llm | parser

# Run
result = chain.invoke({"topic": "AI"})
print(result)
```

# JSON Output Parser

ka use tab hota hai jab hum LLM se output strict JSON format me chahte hain.

# LLM ka text → clean JSON object me convert karna


# Code

```python
from langchain_core.output_parsers import JsonOutputParser
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI

# LLM
llm = ChatOpenAI(model="gpt-4o-mini")

# Prompt
prompt = ChatPromptTemplate.from_template(
    "Extract name, age and city from: {text}"
)

# JSON Output Parser
parser = JsonOutputParser()

# Chain
chain = prompt | llm | parser

# Run
result = chain.invoke({"text": "John is 25 years old and lives in Delhi"})
print(result)
```

# Structured Output Parser

ka use tab hota hai jab hum LLM se output fixed schema (predefined structure) me chahte hain aur usko automatically validated structured data (JSON / dict / model) me convert karna hota hai.

 # LLM ka output → strict format follow karke clean structured data ban jata hai



# Code

```python
from langchain_core.output_parsers import StructuredOutputParser, ResponseSchema
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI

# LLM
llm = ChatOpenAI(model="gpt-4o-mini")

# Step 1: Define schema
schemas = [
    ResponseSchema(name="name", description="Person name"),
    ResponseSchema(name="age", description="Person age"),
    ResponseSchema(name="city", description="Person city")
]

# Step 2: Create parser
parser = StructuredOutputParser.from_response_schemas(schemas)

# Step 3: Prompt with format instructions
format_instructions = parser.get_format_instructions()

prompt = ChatPromptTemplate.from_template(
    "Extract info from text:\n{text}\n{format_instructions}"
)

# Step 4: Chain
chain = prompt | llm | parser

# Step 5: Run
result = chain.invoke({
    "text": "John is 25 years old and lives in Delhi",
    "format_instructions": format_instructions
})

print(result)
```
